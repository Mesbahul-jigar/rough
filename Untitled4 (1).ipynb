## Experimental Work:
1. Analyzing sampling effect

 Simulating and plotting 10 Hz and 110 Hz CT signals for  0 < t < 0.2  s

```python
import numpy as np
import matplotlib.pyplot as plt

# Define frequencies and sampling
F1 = 10
F2 = 110
Fs = 100  # Sampling frequency
Ts = 1 / Fs  # Sampling interval

# Time vector
t = np.arange(0, 0.2, 0.0005)

# Signals
x1t = np.cos(2 * np.pi * F1 * t)
x2t = np.cos(2 * np.pi * F2 * t)

# Plot
plt.figure()
plt.plot(t, x1t, label='10Hz', linewidth=2)
plt.plot(t, x2t, label='110Hz', linewidth=2)

plt.xlabel('continuous time (sec)')
plt.ylabel('Amp')
plt.xlim([0, 0.1])
plt.grid(True)
plt.legend()
plt.title('Two CT CV sinusoids plotted')
plt.show()
```

 Sampling them at 100 Hz and plot in discrete form.

```python
# Discrete time vector
nTs = np.arange(0, 0.2 + Ts, Ts)
n = np.arange(1, len(nTs))  # Not used in plot, but kept to reflect original MATLAB

# Sampled signals
x1n = np.cos(2 * np.pi * F1 * nTs)
x2n = np.cos(2 * np.pi * F2 * nTs)

# Plotting
plt.figure(figsize=(10, 6))

# 10 Hz sampled signal
plt.subplot(2, 1, 1)
plt.stem(nTs, x1n, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.grid(True)
plt.title('10Hz sampled')
plt.xlabel('Discrete time (sec)')
plt.ylabel('Amp')
plt.xlim([0, 0.1])

# 110 Hz sampled signal
plt.subplot(2, 1, 2)
plt.stem(nTs, x2n, basefmt=" ", linefmt='r-', markerfmt='ro')
plt.grid(True)
plt.title('110Hz sampled')
plt.xlabel('Discrete time (sec)')
plt.ylabel('Amp')
plt.xlim([0, 0.1])

plt.tight_layout()
plt.show()
```

  Observing aliasing effects.

```python
plt.figure(figsize=(10, 4))
plt.plot(t, x1t, label='10Hz (cont)')
plt.plot(t, x2t, label='110Hz (cont)')
plt.stem(nTs, x1n, linefmt='r-', markerfmt='ro', basefmt=" ",  label='10Hz sampled')

plt.xlabel('Time (sec)')
plt.ylabel('Amp')
plt.xlim([0, 0.05])
plt.legend()
plt.grid(True)
plt.title('Continuous vs Sampled Signal')
plt.show()

```

2. Analyzing quantization effect:


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fd1 = 1/50
length = 500
n = np.arange(length)

# Simulate the DTCV sinusoid
x = np.cos(2 * np.pi * fd1 * n)

# Number of significant digits to retain
sig_digits = 2  # Change as needed

# Quantization
xq = np.round(x * 10**sig_digits) / 10**sig_digits

# Error signal
error = x - xq

# Power calculations
Px = np.mean(np.abs(x)**2)
Pe = np.mean(np.abs(error)**2)

# SQNR calculation
SQNR = 10 * np.log10(Px / Pe)
print(f"SQNR for {sig_digits} significant digits: {SQNR:.2f} dB")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(n, x, label='Original Signal')
plt.plot(n, xq, '--', label=f'Quantized Signal ({sig_digits} digits)')
plt.plot(n, error, ':', label='Quantization Error')
plt.legend()
plt.title('Quantization Effect on DTCV Sinusoid')
plt.xlabel('Sample Index (n)')
plt.ylabel('Amplitude')
plt.show()
```

```python
fd1 = 1/50
n = np.arange(0, 500)
q = np.arange(0, 11)  # No. of Digits after decimal points to be retained

SQNR = []

for num in range(len(q)):
    x1 = np.cos(2 * np.pi * fd1 * n)
    Px1 = np.sum(np.abs(x1)**2) / len(x1)
    x1q = np.round(x1 * 10**q[num]) / 10**q[num]
    x1e = x1 - x1q
    Pe1 = np.sum(np.abs(x1e)**2) / len(x1e)
    SQNR.append(10 * np.log10(Px1 / Pe1))

plt.figure()
plt.plot(q, SQNR)
plt.xlabel('Significant Digits')
plt.ylabel('SQNR (dB)')
plt.title('SQNR vs Significant Digits')
plt.grid(True)
plt.show()
```
